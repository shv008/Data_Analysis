```python
#changing date time format
df['date_added'] = pd.to_datetime(df['date_added'] , format ='mixed')
```

```python
df['country'].fillna("Unknown" , inplace = True)
```

```python
#removing missing values
df = df.dropna()
```

```python
df.nunique()
```

```python
#Data Cleaning: Dropping Unnecessary Columns
#n this step, we dropped the `show_id` and `description` columns from the original dataset.
#- `show_id` is a unique identifier and not required for our analysis.
#- `description` contains textual data which wonâ€™t be used in the current scope of EDA (but can be used later for NLP or sentiment analysis).
#We also reset the index to ensure the new DataFrame is clean and sequentially indexed.

new_df = df.drop(['show_id' , 'description'] , axis=1)
new_df.reset_index(drop=True , inplace= True)
new_df.head()
```

```python
#ðŸ“ˆ Distribution of Content Release Years
#In this step, we visualized the distribution of `release_year` using Seaborn's `displot()`.
#This helps us understand how the volume of Netflix content has changed over time. As shown in the plot, there has been a sharp increase in releases after 2010, peaking around 2018â€“2020, which aligns with Netflix's global expansion and rise in original productions.

sns.displot(new_df['release_year'])
plt.title('Distribution of Content Release Years on Netflix', fontsize=14)
plt.xlabel('Release Year')
plt.ylabel('Density')
plt.show()
```

```python
#ðŸ”¢ Rating Distribution Analysis

#In this step, we analyzed the distribution of content ratings using `value_counts()` and converted the result into a DataFrame for better readability.

#This shows how Netflix content is classified based on viewing suitability. Key insights:
#- `TV-MA` (Mature Audience) is the most common rating, indicating a large portion of Netflix content is intended for adults.
#- `TV-14` and `R` follow, also targeting older teens and adults.
#- There's relatively less content aimed at younger audiences (`TV-Y`, `G`, etc.).

#This distribution can help in understanding Netflix's target audience and content focus.

ratings = new_df['rating'].value_counts()
ratings_df = pd.DataFrame(ratings)
ratings_df
```

```python
#Count of Titles by Rating â€“ Strip Plot
#In this step, a **strip plot** was used to visualize the distribution of content across different ratings. 
#Each dot represents the number of titles available under a specific rating category.
#This vertical strip plot gives a clean comparative view of the **frequency of ratings** and 
#clearly highlights that `TV-MA`, `TV-14`, and `R` dominate Netflixâ€™s content library.
#While it's visually simple, this plot is best used when showing individual observations. 
#For summarized frequency counts like these, a **bar plot or horizontal bar chart** may offer better clarity

sns.relplot(x='count' , y='rating' , data=ratings_df)
```

```python
countries = new_df['country'].value_counts()
countries
```

```python
#GROUPING
country_grp = new_df.groupby(['country'])
india_df = country_grp.get_group('India')
india_df
```

```python
### ðŸ‡®ðŸ‡³ Content Trend Over Time in India by Type (Movie vs TV Show)

#This **count plot** visualizes the number of Indian titles released on Netflix each year, categorized by type:
#**Movie** and **TV Show**.
#Key observations:
#- There is a **significant rise in content** starting from 2010, with a peak around 2018â€“2019.
#- Movies (blue bars) have consistently outnumbered TV shows (orange bars), indicating Netflixâ€™s stronger focus on 
#movie content in India.
#- TV shows have a much smaller footprint, though their presence has slightly grown in recent years.

#Customizations:
#- The x-axis is adjusted to show every 5th year for better readability using `ticker.MultipleLocator(5)`.
#- X-axis labels are rotated for clarity.

#> ðŸ“Œ **Insight:** This upward trend reflects Netflixâ€™s growing investment in Indian regional and Bollywood content,
#particularly around 2017â€“2019.

g = sns.countplot(x='release_year', hue='type' , data=india_df)
g.xaxis.set_major_locator(ticker.MultipleLocator(5))  # show every 5th year
g.tick_params(axis='x', rotation=45)                  # rotate labels
plt.tight_layout()
plt.show()
```
